```python
# 3.1 - Histogram of active power consumption
import matplotlib.pyplot as plt
import seaborn as sns
df.isnull().sum()
# x = df['Global_active_power'].astype(float) - was showing error because of '?'
```

```python
x = df['Global_active_power'].astype(float)
x.max()
```

```python
plt.figure(figsize=(7,7))
plt.hist(x)
plt.show()
```

```python
# 3.2 - Global active power consumption over time
t = df['Time'].unique()
```

```python
y = df['Time']
plt.figure(figsize=(15,10))
plt.scatter(y,x)
plt.xlabel("Time")
plt.ylabel("Global Power Consumption")
plt.show()
```

```python
y = df['Time']
plt.figure(figsize=(18,9))
# using other parameters because there is a huge number of values - from towardsdatascience article by Paul Gavrikov
plt.scatter(y, x, s=0.4, alpha=0.9) # Here - s controls size of each point and alpha - gives the opacity
plt.xlabel("Time")
plt.ylabel("Global Power Consumption")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 9))
sns.scatterplot(y,x, ax=ax)
```

```python
# 3.3  Energy Sub metering
sm1 = df['Sub_metering_1'].astype(float)
sm2 = df['Sub_metering_2'].astype(float)
sm3 = df['Sub_metering_3'].astype(float)
z = df['Date']
```

```python
# setting the figure size
plt.figure(figsize=(15,25))

# plot several plots at the same time
plt.subplot(311) # 1 - one row, 3 - three columns, 1 - first position
plt.title("Graph of Sub-metering 1 against Date")
plt.scatter(z, sm1)

plt.subplot(312)  # 1- one row, 3 - three columns, 2 - second position
plt.title("Graph of Sub-metering 2 against Date")
plt.scatter(z, sm2)

plt.subplot(313)
plt.title("Graph of Sub-metering 3 against Date")
plt.scatter(z, sm3)

plt.show()
```
